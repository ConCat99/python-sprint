### 4. Aggregatoin with GroupBy

```python
# Average sales before & after - group by Customer Seg
avg_sales = df.groupby("Customer_Segment").agg(
    avg_before_sales = ("Sales_Before", "mean"), 
    avg_after_sales = ("Sales_After", "mean")
)
# avg_sales.head()

# Satisfaction improvement by Group (Control vs Treatment)
sum_sat = df.groupby("Group").agg(
    sum_before_satisfaction = ('Customer_Satisfaction_Before', 'sum'), 
    sum_after_satisfaction = ('Customer_Satisfaction_After', 'sum')
).reset_index()
# print('sum_sat >>', sum_sat)

satis_diff = sum_sat['sum_after_satisfaction'] - sum_sat['sum_before_satisfaction']
# print('satis_diff >>>', satis_diff)

#Find which Customer_Segment had the biggest sales increase

sales_diff = df["Sales_After"] - df["Sales_Before"] # sales diff per record
# sales_diff.head()

# customer_sales_diff = df.groupby("Customer_Segment").agg(

# )

```
