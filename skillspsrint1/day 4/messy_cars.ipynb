```python
import pandas as pd
```

```python
df = pd.read_csv("messy_cars.csv")
df.head(5)
```

### Tasks: 

- Drop or fill missing values appropriately. <br>
- Filter the dataset for a meaningful subset. <br>
- Sort the data by miles descending.<br>
- Rename columns for clarity.<br>
- Save the cleaned dataset as cars_cleaned.csv.<br>

```python
# Drop or fill missing values appropriately -> drop cars with missing miles
dropped_df = df.dropna()
df.head(5)

# Fill cars with no cost as "TBD"
df["Money"] = df["Money"].fillna("TBD")
df.head(5)
```

```python
# Sort the data by miles descending
miles_dsc = df.sort_values(by = ["Miles"], ascending = False)
miles_dsc.head(5)
```

```python
# Rename columns for clarity.
df.columns= ["Car Registeration", "Manufacturer", "Model", "Date_Year", "Current_Value", "Mileage", "Colour"]
df.head(5)
```

```python
df.to_csv("semi_cleaned_cars.csv", index=False)
#to_pickle?
```
