```python
import pandas as pd
```

```python
df = pd.read_csv("students_data_messy.csv")
df.head()
```

```python
## drop rows where any value is missing
dropped_df = df.dropna().to_string()
print(dropped_df)

## drop rows where a specific column is missing
dropped_df_col = df.dropna(subset=["Age"])
dropped_df_col.head()
```

```python
# fill the missing values with a constant -> This means that the age is filled as 0, may not be as useful as it may not be meaningful data
df["Age_Filled"] = df["Age"].fillna(0)


#fill missing vlaues with a mean or median or statistic-> Setting the mean score if it's empty
df["Score_Filled"] = df["Score"].fillna(df["Score"].mean())
df.head(20)
```

```python
##Filtering & Sorting

# filter multiple conditions use and or, age and score, attendance and extracurricular

sorted_sixteens_yo = df[(df["Age"] == 16]) & [(df.sort_values(by = "Attendance", ascending= True)]

print(sorted_sixteen_yo)
# Not sure what's happening with brackets^^
```

```python
# sort by score ascending

sort_asc = df.sort_values(by = "Score", ascending=True)
sort_asc.head(5)

# sort by score descending
sort_dsc = df.sort_values(by = "Score", ascending=False)
sort_dsc.head(5)

# sort by age asc & score desc
sorted_2 = df.sort_values(by = ["Score","Age"], ascending=[False,True])
sorted_2.head(5)

```

```python
## Renaming Columns

# rename specific columns
df_renamed = df.rename(columns = {"Extracurricular": "After School Club"})
df_renamed.head()
```
